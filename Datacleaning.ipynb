## Different method to dealing with missing data

```python
df=pd.read_csv("googleplaystore.csv")
df
```

```python
df.isnull().sum()
```

## Deleting the column with missing data

```python
df1=df.copy()
df1.drop("Content Rating", axis=1 , inplace=True)
```

```python
df1.isna().sum()
```

## Deleting row with missing data

```python
df2=df.copy()
df2.dropna( inplace=True)
```

```python
df2.shape
```

## Filling the missing values- Imputation

Filling Numeric values

```python
df3=df.copy()
mean_value=df3["Rating"].mean()
df3["Rating"].fillna( mean_value ,inplace=True)
df3.sample(10)
df3["Rating"].isna().sum()
```

filling categorical data

```python
df4=df.copy()
df4["Type"].fillna(df["Type"].value_counts().index[0],inplace=True)
```

```python
df4.isna().sum()
```

making another category to fill values

```python
df5=df.copy()
df5["Type"].fillna("not_given",inplace=True)
```

```python
df5.sample(10)
```

## Advanced Imputation

```python
df6=df.copy()
df6["Rating"].interpolate(inplace=True)
```

```python
df6["Rating"].isnull().sum()
```

## Outlier

```python
import pandas as pd
import numpy as np 
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("Scholarship.csv")
```

```python
df=data.copy()
df
```

```python
df.shape
```

```python
plt.subplot(1,2,1)
sns.distplot(df["semester_percentage"])	
plt.subplot(1,2,2)
sns.distplot(df["scholarship_exam_marks"])
plt.show()
```

```python
mean=df["semester_percentage"].mean()
std=df["semester_percentage"].std()
print(mean, "," ,std)
```

## finding outliers

```python
highest_allowed=mean+3*std
lowest_allowed=mean-3*std
print(highest_allowed, "," ,lowest_allowed)
```

## Trimming

```python
df[(df["semester_percentage"]<lowest_allowed )| (df["semester_percentage"]>highest_allowed)]
```

```python
df[(df["semester_percentage"]>lowest_allowed )& (df["semester_percentage"]<highest_allowed)]
```

## Finding z-score

```python
df["z-score"]=(df["semester_percentage"]-df["semester_percentage"].mean())/df["semester_percentage"].std()
df.head()
```

```python
df[(df["z-score"]<-3)| (df["z-score"]>3)]
```

```python
df[(df["z-score"]>-3)& (df["z-score"]<3)]
```

## Capping

```python
upper_limit=mean+3*std
lower_limit=mean-3*std
```

```python
df["semester_percentage	"]= np.where(df["semester_percentage"]>upper_limit,upper_limit,
                                     np.where(df["semester_percentage"]<lower_limit,lower_limit,df["semester_percentage"]))
```

```python
df
```

```python
df[(df["semester_percentage"]<lowest_allowed )& (df["semester_percentage"]>highest_allowed)]
```

## Outlier Removal Using IQR

```python
data
```

```python
sns.distplot(data["scholarship_exam_marks"])
plt.show()
```

```python
sns.boxplot(data["scholarship_exam_marks"])
plt.show()
```

```python
Q1=data["scholarship_exam_marks"].quantile(0.25)
Q3=data["scholarship_exam_marks"].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
data["scholarship_exam_marks"].describe()
```

```python
max=Q3+1.5*IQR
min=Q1-1.5*IQR
```

```python
max
```

```python
min
```

```python
data[(data["scholarship_exam_marks"]>max) | (data["scholarship_exam_marks"]<min)] # finding outliers
```

## Trimming

```python
new_data=data[(data["scholarship_exam_marks"]<max) & (data["scholarship_exam_marks"]>min)] # removing outliers
```

```python
new_data
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(data["scholarship_exam_marks"])

plt.subplot(2,2,2)
sns.boxplot(data["scholarship_exam_marks"])

plt.subplot(2,2,3)
sns.distplot(new_data["scholarship_exam_marks"])

plt.subplot(2,2,4)
sns.boxplot(new_data["scholarship_exam_marks"])
plt.show()
```

## Capping

```python
data["scholarship_exam_marks"]=np.where(data["scholarship_exam_marks"] >max,max,
                                       np.where (data["scholarship_exam_marks"]<min, min ,data["scholarship_exam_marks"]))
```

```python
data
```

```python
sns.boxplot(data["scholarship_exam_marks"])
```

## Outlier Removal Using Percentile

```python
df=pd.read_csv("weight-height.csv")
df.head()
```

```python
sns.distplot(df["Height"])
```

```python
sns.boxplot(df["Height"])
```

```python
min_limit=df["Height"].quantile(0.01)
max_limit=df["Height"].quantile(0.99)
```

```python
min_limit
```

```python
max_limit
```

```python
new_data=df[(df["Height"]<min_limit) | (df["Height"]> max_limit)] # finding outliers
```

```python
new_data
```

## Trimming

```python
new=df[(df["Height"]>min_limit)&(df["Height"]<max_limit)]
```

```python
new
```

```python
sns.boxplot(new["Height"])
```

## Capping

```python
df["Height"]=np.where(df["Height"]<min_limit,min_limit,np.where(df["Height"]>max_limit,max_limit,df["Height"]))
```
